# Data Visualization

```python
plt.figure(figsize=(20,12))
sns.barplot(data= train,
           x= "Married/Single",
           y= "Income");
plt.style.use("ggplot")
plt.xlabel("Married/Single", fontsize=25);
plt.ylabel("Income", fontsize=25);
plt.tick_params(which= "major", axis= "x", labelsize= 15)
plt.tick_params(which= "major", axis= "x", labelsize= 15)
```

```python
plt.figure(figsize=(20,12))
sns.histplot(data= train,
           x= "House_Ownership",
            color= ["red"])
plt.xlabel("House_Ownership", fontsize= 15)
plt.ylabel("Count", fontsize= 15)
plt.tick_params(which= "major", axis= "x", labelsize= 15)
plt.tick_params(which= "major", axis= "y", labelsize= 15)
```

```python
plt.figure(figsize=(20,12))
sns.histplot(data= train,
           x= "Car_Ownership",
            color= "blue")
plt.xlabel("House_Ownership", fontsize= 15)
plt.ylabel("Count", fontsize= 15)
plt.tick_params(which= "major", axis= "x", labelsize= 15)
plt.tick_params(which= "major", axis= "y", labelsize= 15)
```

```python
plt.figure(figsize=(20,12))
sns.barplot(data= train,
           x= "Age",
           y= "Income");
plt.style.use("ggplot")
plt.xlabel("Age", fontsize=25);
plt.ylabel("Income", fontsize=25);
plt.tick_params(which= "major", axis= "x", labelsize= 15)
plt.tick_params(which= "major", axis= "x", labelsize= 15)
```
